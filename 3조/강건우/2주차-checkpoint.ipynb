```python
def pltpie(label, sample):
    data = Train[Train[label] == sample]
    data_c = data['credit'].value_counts()
    data_c
    plt.pie(x=data_c, labels=data_c.index, autopct='%1.1f%%')
    plt.title(label+'_'+str(sample))
    plt.show()
```
